```python
import pandas as pd
import matplotlib.pyplot as plt

#filepath= './RNAseq/A/A_1.fq/A_1.fq'
filepath= 'ERR037900_1.first1000.fastq'

def readFastQ(filepath):
    sequences= []
    qualities=[]
    with open(filepath) as f:
        while True:
            f.readline()
            seq= f.readline().rstrip()
            #print(seq)
            f.readline()
            qual= f.readline().rstrip()
            if len(seq) == 0:
                break
            sequences.append(seq)
            qualities.append(qual)
    return sequences, qualities

def phred33ToQ(phredQual):
    #convert from "phred33" to Unicode format; one character at a time
    return ord(phredQual) - 33  #ord gets unicode
    
seqs, quals= readFastQ(filepath)
#print(sequence)
print(quals[:5])

def hist(qualList):
    #make histogram of quality scores
    histL= [0]*50
    for i in range(50):
        histL[i]= 0
    for i in qualList:
        for j in i:
            qualNum= phred33ToQ(j)
            histL[qualNum] += 1
    return(histL)

scoreHist= hist(quals)
# print histogram values
print('scores:', scoreHist)

plt.bar(range(50),scoreHist)
plt.show()
```

```python
def GCbyPos(reads):
    #get GC fraction by pos in read
    GC_count= [0]*100 #assuming max length is 100
    total_count= [0]*100
    for seq in reads:
        for pos in range(len(seq)):
            if seq[pos] in ['G','C']:
                GC_count[pos] += 1
            total_count[pos] += 1

    frac= [0]*100
    for pos in range(len(GC_count)):
        if total_count[pos]>0:
            frac[pos] = GC_count[pos]/float(total_count[pos])

    return frac, GC_count, total_count

GC_fractions, GCs, totals= GCbyPos(seqs)
print('GC fractions', GC_fractions)

plt.plot(range(len(GC_fractions)),GC_fractions)
plt.title('GC-fraction')
plt.show()

plt.plot(range(len(totals)),totals)
plt.title('Totals')
plt.show()
```

```python
def naive(p,t):
    #naive exact alignment: find pattern p in text t
    #like "find" function
    matches_firstPos= []
    for firstPos in range(len(t)-len(p)+1):
        match= True
        for j in range(len(p)):
            if p[j] != t[firstPos+j]:
                match= False
                break;
    matches_firstPos.append(firstPos)
    return(matches_firstPos)    

#test1
t1= 'GGACGG'
p1= 'AC'  #expected: match at 2
print( naive(p1,t1) )
```
